```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import random
```

```python
from google.colab import files
train=pd.read_csv(next(iter(files.upload().keys())))
```

```python
train['sex'].replace({'F':0,'M':1},inplace=True)
train['is_smoking'].replace({'YES':0,'NO':1},inplace=True)
train['TenYearCHD'].replace({0:-1},inplace=True) #is this necessary?
```

```python
#spliting train dataset into input and ouput
y=train['TenYearCHD'].copy()

#droping highly correlated features
x=train.drop(['BPMeds','prevalentHyp','sysBP','TenYearCHD','id','is_smoking'],axis=1).copy()
```

```python
b=np.array([random.gauss(0,1)for _ in range(len(x.columns))])
```

```python

for num in range(len(x)):
  x1=x.iloc[num]
  y1=y.iloc[num]
  if y1>0 and np.dot(np.transpose(b),x1) <0:#?
    b=b+x1
  if y1 <0 and np.dot(np.transpose(b),x1) >0:
    b=b-x1
count=0
for num in range(len(x)):
  x1=x.iloc[num]
  y1=y.iloc[num]
  if y1>0 and np.dot(np.transpose(b),x1) <0:#?
    count+=1
  if y1 <0 and np.dot(np.transpose(b),x1) >0:
    count+=1

display(100-(len(x)-count)/100)
```
