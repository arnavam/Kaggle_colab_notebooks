```python
from bs4 import BeautifulSoup
import requests
```

```python
with open ('blah.htm','r') as hfile:
  con=hfile.read()
  soup=BeautifulSoup(con,'lxml')
  print(soup.prettify())
  tag=soup.find('h5')#only find first 'h5' file
  print(tag)
  tags=soup.find_all('h5')#find all the files with j'h5'
  for course in tags:
    print(course.text)#take only text

  cards=soup.find_all('div',class_='card')
  for course in cards:
    print(course.h5.text)
    price=course.a.text.split()[-1]
    print(price)
```

```python
p=requests.get('https://www.timesjobs.com/candidate/job-search.html?searchName=recentSearches&from=submit&luceneResultSize=25&txtKeywords=python&postWeek=60&searchType=personalizedSearch&actualTxtKeywords=python&searchBy=0&rdoOperator=OR&gadLink=python').text
print(p)
soup=BeautifulSoup(p,'lxml')
jobs=soup.find('li',class_='clearfix job-bx wht-shd-bx')
print(jobs)
```
