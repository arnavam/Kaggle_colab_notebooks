```python
# @title finding categorical & continous feature


# Define a function to find categorical and continuous features in a dataset
def cat_fea(train):
    # Get the number of unique values for each feature in the dataset
    unique_counts = train.nunique()

    # Threshold to distinguish continuous and categorical features
    threshold = 12

    # Identify continuous variables as those with more unique values than the threshold
    continuous_vars = unique_counts[unique_counts > threshold].index.tolist()

    # Identify categorical variables as those with unique values less than or equal to the threshold
    categorical_vars = unique_counts[unique_counts <= threshold].index.tolist()

    # Return the lists of continuous and categorical variables
    return continuous_vars, categorical_vars

# Call the function on the dataset, excluding object type columns
cont, cat = cat_fea(data.select_dtypes(exclude=['object']))
```

```python
# @title category feature disturbutions

def target_plot(column, train):

    # Define a custom color palette
    custom_palette = sns.color_palette("pastel")

    # Create a figure with two subplots
    fig, ax = plt.subplots(1, 2, figsize=(14, 6))

    # Plot a pie chart of the value counts for the column
    train[column].value_counts().plot.pie(autopct='%2.1f%%', ax=ax[0], shadow=True, colors=custom_palette)
    ax[0].set_ylabel(f'{column}')

    # Plot a bar plot of the value counts for the column
    sns.countplot(x=column, data=train, ax=ax[1], palette=custom_palette, hue=column)

    # Set titles for the subplots
    ax[0].set_title(f'{column} Distribution (Pie Chart)', fontsize=14)
    ax[1].set_title(f'{column} Count (Bar Plot)', fontsize=14)

    # Set the overall title for the figure
    plt.suptitle(f'{column} Visualization', fontsize=16, fontweight='bold')

    # Add gridlines to the bar plot
    plt.grid(axis='y', linestyle='--', alpha=0.5)

    # Show the plot
    plt.show()

    print("Columns in Target column are:", train[column].value_counts())

for i in cat:
    target_plot(i, data)

#This means that most  of the data we have is from movies in 2016
```

```python
# @title top 10 movies overall(based on rating and metascore)

# Create a new column 'm' by scaling the 'Metascore' column to a 0-10 range
data['m'] = data['Metascore'] / 10

# Create a new column 'combined rating' by averaging the 'Rating' and 'm' columns
data['combined rating'] = data[['Rating', 'm']].mean(axis=1)

# Sort the DataFrame by the 'combined rating' column in descending order and display the top 10 rows
data.sort_values('combined rating', ascending=False).head(10)


#This is not the only method like for example you could try to find max of both
#If you think that people are most likely to give high score than low score, the you could find min of both or
#you could even combine all this method and find the film that always comes up top or relatively top list.
```

```python
# @title top 10 movies overall(based on rating and metascore and votes)

# Mean rating across all movies
C = data['Rating'].mean()

# Minimum number of votes required to be considered
m = data['Votes'].quantile(0.70)

# Function to compute weighted rating
def weighted_rating(x, m=m, C=C):
    v = x['Votes']
    R = x['combined rating']
    return (v / (v + m) * R) + (m / (v + m) * C)

# Apply the function to compute weighted rating
data['Weighted Rating'] = data.apply(weighted_rating, axis=1)

# Sort movies based on weighted rating
data.sort_values(by='Weighted Rating', ascending=False).head(10)

#Here we used a general equation to combine the three features
```

```python
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt

# Initialize lists to store individual actors and their casts
Cast = []
casts = []

# Loop through each entry in the 'Actors' column
for i in data['Actors']:
    # Check if the entry is a string
    if type(i) == str:
        # Split the actors by comma and add to the 'casts' list
        casts.append(i.split(','))
        # Add each actor to the 'Cast' list
        for x in i.split(','):
            Cast.append(x)

# Count the occurrences of each actor
dic = Counter(Cast)

# Create a DataFrame to store actors and their counts, and sort by count in descending order
df_Cast = pd.DataFrame(data={'Cast': list(dic.keys()), 'Count': list(dic.values())}).sort_values(by='Count', ascending=False)

# Set the 'Cast' column as the index of the DataFrame
df_Cast.index = df_Cast.Cast

# Plot the top 10 actors/actresses by the number of movies released
df_Cast.Count.head(10).plot.bar(figsize=(12, 6), color=["r", "b", "g"])
plt.title("Top 10 Actor/Actress")
plt.xlabel("Cast")
plt.ylabel("Number of Movies Released")
plt.show()

# Add the 'casts' list to the original data DataFrame
data['casts'] = casts
```

```python
# @title Converting Genres and casts to into multiple binary columns
Data=data.copy()
def encode(Data):
  l=Data.groupby('Director')['Weighted Rating'].mean().sort_values(ascending=False)
  Data['Director'].replace(l.round(1),inplace=True)

  # Perform one-hot encoding on the expanded DataFrame
  one_hot_encoded_genres = pd.get_dummies(Data.explode('Genres')['Genres'])

  # Aggregate the one-hot encoded columns back to the original DataFrame
  Data= Data.drop(columns=['Genres']).join(one_hot_encoded_genres).groupby(level=0).max()


  one_hot_encoded_genres = pd.get_dummies(Data.explode('casts')['casts'])

  # Aggregate the one-hot encoded columns back to the original DataFrame
  Data= Data.drop(columns=['casts']).join(one_hot_encoded_genres).groupby(level=0).max()
  return Data
Data=encode(Data)
```

```python
# @title Training ml model to predict missing values
from sklearn.model_selection import train_test_split

# Define the target variable
y = Data['Revenue (Millions)']

# Drop the target variable from the feature set and exclude non-numeric columns
x = Data.drop('Revenue (Millions)', axis=1).select_dtypes(exclude=['object'])

# Split the data into training and test sets
xtr, xte, ytr, yte = train_test_split(x, y, test_size=0.2, shuffle=True)

from sklearn.ensemble import RandomForestRegressor

# Initialize the Random Forest Regressor with 100 trees
model = RandomForestRegressor(n_estimators=100)

# Fit the model on the training data
model.fit(xtr, ytr)

# Evaluate the model on the test data and return the R^2 score
model.score(xte, yte)
```

```python
# @title Updating nf and data with new values

# Predict the 'Revenue (Millions)' for the data in `nf`
pre = model.predict(nf.drop('Revenue (Millions)', axis=1).select_dtypes(exclude=['object']))

# Update the 'Revenue (Millions)' column in `nf` with the predicted values
nf.loc[nf.index, 'Revenue (Millions)'] = pre

# Update the 'Revenue (Millions)' column in `data` with the predicted values
data.loc[nf.index, 'Revenue (Millions)'] = pre

# Display information about the 'Revenue (Millions)' column in `data`
data['Revenue (Millions)'].info()
```
